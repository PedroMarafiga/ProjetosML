## para tentar prever o salário, dividiremos em X e Y nosso dataset

```python
man_salaryX = man_salary.loc[:,['school','exper', 'union',"maried","health","industry","occupation"]].values
```

```python
man_salaryX
```

```python
man_salaryY = man_salary.loc[:, ["wage"]].values
```

```python
man_salaryY
```

## devemos etratar os atributos categóricos utilizando LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
label_encoder_teste = LabelEncoder()
```

```python
man_salaryX[1,:]
```

```python
label_encoder_union = LabelEncoder()
label_encoder_maried = LabelEncoder()
label_encoder_health = LabelEncoder()
label_encoder_industry = LabelEncoder()
label_encoder_occupation = LabelEncoder()
```

```python
man_salaryX[:,2] = label_encoder_union.fit_transform(man_salaryX[:,2])
man_salaryX[:,3] = label_encoder_maried.fit_transform(man_salaryX[:,3])
man_salaryX[:,4] = label_encoder_health.fit_transform(man_salaryX[:,4])
man_salaryX[:,5] = label_encoder_industry.fit_transform(man_salaryX[:,5])
man_salaryX[:,6] = label_encoder_occupation.fit_transform(man_salaryX[:,6])
```

```python
man_salaryX[1,:]
```

## agora para fazer o OneHotEncoder, serve para não dar "vantagem" para nenhuma das váriaveis categóricas

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
onehotencoder_salary = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [2,3,4,5,6])], remainder='passthrough')
```

```python
man_salaryX = onehotencoder_salary.fit_transform(man_salaryX).toarray()
```

```python
man_salaryX
```

```python
man_salaryX.shape
```

# Escalonar

```python
from sklearn.preprocessing import StandardScaler
scaler_salary = StandardScaler()
man_salaryX = scaler_salary.fit_transform(man_salaryX)
```

# Divisão das bases em treinamento e teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_salary_treinamento, X_salary_teste, y_salary_treinamento, y_salary_teste = train_test_split(man_salaryX, man_salaryY, test_size = 0.2, random_state = 0)
```
